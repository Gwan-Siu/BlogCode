```python
import numpy as np
import skimage.transform as st
import matplotlib.pylab as plt
%matplotlib inline
```

## Hough Transform

```python
img = np.zeros((101,101)) # create background image
indx = np.arange(25,75)
img[indx[::-1], indx]=255 # line
img[indx,indx]=255 #line
img[:,78]=255
img[43,:]=255
h, theta, d = st.hough_line(img)
```

```python
h, theta, d = st.hough_line(img)

plt.figure(figsize=(15,20))
plt.subplot(131)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(132)
plt.imshow(np.log(1+h))
plt.title('Hough transform')
plt.axis('off')

plt.subplot(133)
plt.imshow(img,cmap='gray')
row, col=img.shape
for _, angle, dist in zip(*st.hough_line_peaks(h, theta, d)):
    y0 = (dist - 0 * np.cos(angle)) / np.sin(angle)
    y1 = (dist - col * np.cos(angle)) / np.sin(angle)
    plt.plot((0, col), (y0, y1), '-r')
plt.axis((0, col, row, 0))
plt.title('detected line')
plt.axis('off')
plt.show()
```
